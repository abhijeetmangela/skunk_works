```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
file_path = "T1-18_3 m_s-LLT_plot.xlsx"
df = pd.read_excel(file_path)

# Clean column names
df.columns = [str(c).strip() for c in df.columns]

# Extract required columns
alpha = pd.to_numeric(df["α (°)"], errors='coerce')
cm = pd.to_numeric(df["Cm"], errors='coerce')

# Clean data
data = pd.DataFrame({"alpha": alpha, "cm": cm}).dropna()
data = data.sort_values(by="alpha")

# Compute derivative (Cm_alpha)
cm_alpha = np.gradient(data["cm"], data["alpha"])

# ---- Plot Cm vs alpha ----
plt.figure()
plt.plot(data["alpha"], data["cm"], marker='o')
plt.xlabel(r'$\alpha$ (deg)')
plt.ylabel(r'$C_m$')
plt.title(r'$C_m$ vs $\alpha$')
plt.grid()

# ---- Plot Cm_alpha vs alpha ----
plt.figure()
plt.plot(data["alpha"], cm_alpha, marker='o')
plt.xlabel(r'$\alpha$ (deg)')
plt.ylabel(r'$C_{m_\alpha}$')
plt.title(r'$C_{m_\alpha}$ vs $\alpha$')
plt.grid()
plt.savefig("cm_alpha_alpha.png")

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
df = pd.read_excel("T1-18_3 m_s-LLT_plot.xlsx")
df.columns = [str(c).strip() for c in df.columns]

# Extract and clean
alpha = pd.to_numeric(df["α (°)"], errors='coerce')
cm = pd.to_numeric(df["Cm"], errors='coerce')

data = pd.DataFrame({"alpha": alpha, "cm": cm}).dropna()
data = data.sort_values(by="alpha")

# Compute derivative
cm_alpha = np.gradient(data["cm"], data["alpha"])

# Target alpha
alpha_target = 3.0
idx = (np.abs(data["alpha"] - alpha_target)).argmin()

alpha_val = data["alpha"].iloc[idx]
C_m_AC_w = cm_alpha[idx]

# Plot
plt.figure()
plt.plot(data["alpha"], cm_alpha, marker='o', label=r'$C_{mACW_\alpha}$')

plt.axvline(x=alpha_val, linestyle='--', label=r'$\alpha = 3^\circ$')
plt.scatter(alpha_val, C_m_AC_w)

# plt.text(alpha_val, C_m_AC_w,
#          f'{C_m_AC_w:.4f}',# f'   $C_{{m\\_AC\\_w}}$ = {C_m_AC_w:.4f}',
#          verticalalignment='bottom')

plt.xlabel(r'$\alpha$ (deg)')
plt.ylabel(r'$C_{m\_ACw\_\alpha}$')
plt.title(r'$C_{m\_ACW\_\alpha}$ at $\alpha = 3^\circ$')

plt.grid()
plt.legend()
plt.tight_layout()

plt.show()

print(f"C_m_AC_w at alpha = {alpha_val:.1f}° is {C_m_AC_w:.5f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
df = pd.read_excel("T1-18_3 m_s-LLT_plot.xlsx")
df.columns = [str(c).strip() for c in df.columns]

# Extract and clean
alpha = pd.to_numeric(df["α (°)"], errors='coerce')
cm = pd.to_numeric(df["CL"], errors='coerce')*np.pi*2

data = pd.DataFrame({"alpha": alpha, "cm": cm}).dropna()
data = data.sort_values(by="alpha")

# Compute derivative
cm_alpha = np.gradient(data["cm"], data["alpha"])

# Target alpha
alpha_target = 3.0
idx = (np.abs(data["alpha"] - alpha_target)).argmin()

alpha_val = data["alpha"].iloc[idx]
C_m_AC_w = cm_alpha[idx]

# Plot
plt.figure()
plt.plot(data["alpha"], cm_alpha, marker='o', label=r'$C_{Lw_\alpha}$')

plt.axvline(x=alpha_val, linestyle='--', label=r'$\alpha = 3^\circ$')
plt.scatter(alpha_val, C_m_AC_w)

plt.text(alpha_val, C_m_AC_w,
         f'   $C_{{L\\_alpha}}$ = {C_m_AC_w:.4f}',
         verticalalignment='bottom')

plt.xlabel(r'$\alpha$ (deg)')
plt.ylabel(r'$C_{L\_w\_\alpha}$')
plt.title(r'$C_{L\_w\_\alpha}$ at $\alpha = 3^\circ$')

plt.grid()
plt.legend()
plt.tight_layout()

plt.show()

print(f"C_m_AC_w at alpha = {alpha_val:.1f}° is {C_m_AC_w:.5f}")
```
